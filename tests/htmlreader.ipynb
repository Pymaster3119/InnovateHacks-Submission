```python
resp = requests.get(urls[1])
```

```python
html = resp.text
```

```python
len(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
tree = lxml.html.fromstring(html)
```

```python
tags = defaultdict(list)
for element, attribute, link, pos in tree.iterlinks():
  tags[element.tag].append(link)
```

```python
tags.keys()
```

```python
for form in tree.forms:
  for elem in form:
    print(elem)
```

```python
len(tree.forms)
```

```python
for form in tree.forms:
  print(form)
  for inp in form.inputs:
    print(inp,inp.attrib,inp.value,inp.getparent())
```

```python
def drop_extra(root):
  if root is None:
    return
  print(root.tag)
  if root.tag not in ['html', 'body', 'a', 'form', 'input', 'submit', 'p', 'li', 'ul', 'h1', 'h2', 'h3', 'h4', 'b']:
    try:
      root.drop()
    except AttributeError:
      pass
    return
  for c in root.getchildren():
    drop_extra(c)
```

```python
drop_extra(tree)
```
